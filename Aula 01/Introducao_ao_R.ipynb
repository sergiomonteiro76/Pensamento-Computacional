**Exemplo 09**

```r
install.packages("ggpubr")
install.packages("tidyverse")

library(tidyverse)
library(ggpubr)
theme_set(theme_pubr())

#Limpar ambiente de desenvolvimento
rm(list = ls())

# variaveis dependentes/preditoras
x <- c(10, 15, 13, 18, 14, 15, 15, 16, 21, 13)

# vetor saida/resposta
y <- c(3, 5, 7, 9, 13, 15, 17, 19, 21, 23)


dados <- data.frame(fresposta = y,
                    vindependente = x)


# Aplicar a função de regressão linear
fregressaoLinear <- lm(fresposta~vindependente,data = dados)

# Predição para x = 25.
a <- data.frame(vindependente = 25)
resultado <-  predict(fregressaoLinear,a)
print(resultado)


#Imprimir Regressão Linear
ggplot(dados, aes(vindependente, fresposta)) +
       labs(title = 'Exemplo Regressão Linear',
       y = 'Função',
       x = 'independentes')+
  geom_point(colour = "red", size = 3) +
  stat_smooth(method = lm)

#BoxPlot
boxplot(fresposta~vindependente, data = dados)
abline(fregressaoLinear)
```

**Exemplo 10**

```r
# Dados
x1 <- c(1, 2, 3, 4, 5)
x2 <- c(0.5, 1.5, 2.5, 3.5, 4.5)
y <- c(3, 6, 7, 8, 10)

# Regressao Linear Multipla
model <- lm(y ~ x1 + x2)

# Organizar os dados para imprimir no plano
x1_grid <- seq(min(x1), max(x1), length.out = 20)
x2_grid <- seq(min(x2), max(x2), length.out = 20)
grid <- expand.grid(x1 = x1_grid, x2 = x2_grid)

# Predicao usando o modelo
grid$y_pred <- predict(model, newdata = grid)

# Imprimir resultado no plano
install.packages("scatterplot3d")
library("scatterplot3d")
s3d <- scatterplot3d(x1, x2, y, color = "blue", main = "Exemplo de Regressao Linear Multipla", xlab = "X1", ylab = "X2", zlab = "Y")
s3d$plane3d(model, draw_polygon = TRUE, color = "red", alpha = 0.2)
```
